## Question 2.a

```python
#assign diiferent value to a and b in range
a2=np.arange(start=-10,stop = 10, step =0.1) 
b2=np.arange(start=-10,stop = 10, step =0.1)
```

## Question 2.b 

```python
# Defined the function for element error and average error
def ele_err(a,b):
    y2 = a*x+b
    e1 = y1-y2
    E = sum(e1**2)/100
    return E, e1
```

```python
# Computing element error for random value for a and b
com_error = ele_err(a,b)[1]
print("computed element error is :\n",com_error)
```

## Question 2.c

```python
# computing average error for random value of a and b
avg_error = ele_err(a,b)[0]
print("average error for random value of a and b is : \n",avg_error)
```

## Question 2.d

```python
# Avg Error for all combination of a and b
error = []
x_plot = []
y_plot = []
for i in range(len(a2)):
    for j in range(len(b2)):
        err = ele_err(a2[i],b2[j])[0]
        x_plot.append(a2[i])
        y_plot.append(b2[j])
        error.append(err)
print("Average error for all combination of a and b is : \n", error)
```

## Question 2.e

```python
# 3d plot for average error 
from mpl_toolkits.mplot3d import Axes3D 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_plot, y_plot, error, c = error);
```
